Construa agora um ou vários programas Python para processar o texto 'processos.txt' (procurar o ficheiro no Bb) com o intuito de calcular frequências de alguns elementos (a ideia é utilizar arrays associativos, dicionários em Python, para o efeito) conforme solicitado a seguir:

    a) Calcula a frequência de processos por ano (primeiro elemento da data);

    b) Calcula a frequência de nomes próprios (o primeiro em cada nome) e apelidos (o ultimo em cada nome) por séculos e apresenta os 5 mais usados;

    c) Calcula a frequência dos vários tipos de relação: irmão, sobrinho, etc.;

    d) Converta os 20 primeiros registos num novo ficheiro de output mas em formato Json.


```python
import re
import matplotlib.pyplot as plt
from collections import Counter,OrderedDict
import math

lineRegex=re.compile(r"(?P<pasta>\d+)::(?P<data>\d\d\d\d-\d\d-\d\d)::(?P<nome>[^:]*)::(?P<pai>[^:]*)::(?P<mae>[^:]*)::(?P<obs>[^:]*)::")
data=[]
with open("processos.txt",'r') as file:
    for line in file.readlines():
        m = lineRegex.match(line)
        if m:
            data.append(m.groupdict())
```

A)

```python
for line in data:
    ano,mes,dia=line['data'].split('-')
    line['ano']=ano
    line['mes']=mes
    line['dia']=dia
Counter(map(lambda x: x['ano'],data))
```

B)

```python
#5 mais usados
c=Counter(map(lambda x:x['nome'],data))
c.most_common(5)
primeiroNome={}
ultimoNome={}
for line in data:
    seculo = math.ceil(int(line['ano'])/100)
    nomes = line['nome'].split(' ') 
    if seculo not in primeiroNome:
        primeiroNome[seculo]=[]
    if seculo not in ultimoNome:
        ultimoNome[seculo]=[]
    primeiroNome[seculo].append(nomes[0])
    ultimoNome[seculo].append(nomes[-1])
primeiroNome=sorted(primeiroNome.items(), key=lambda x: x[0])
for k,v in primeiroNome:
    print("seculo",k)
    print("primeiro nome:",*Counter(v).most_common(5))
    print("ultimo nome:",*Counter(ultimoNome[k]).most_common(5))

    
      
    
    
    
    
```

C)
Calcula a frequência dos vários tipos de relação: irmão, sobrinho, etc.;

```python
r = re.compile(r",(Irmao|Irma|Sobrinho|Sobrinha|Pai|Mae|Filho|Filha|Tio|Tia)\s")
l=[]
for obs in map(lambda x:x['obs'],data):
    j=r.findall(obs)
    if j!=[]:
        l+=j
Counter(l)
        
```
